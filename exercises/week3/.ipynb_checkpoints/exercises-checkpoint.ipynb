## Interpolation of images, discontinuities

Part (a): Let
\begin{equation*}
    u : \Omega \rightarrow [0,1] \subset \mathbb{R}
\end{equation*}
be an (idealized) continuous image defined in the domain $\Omega := [0,1]^2 \subset \mathbb{R}^2$. We assume the partial derivatives of $u$ to be bounded, i.e.
\begin{align*}
	| \partial_x u | ~&\leq~ c_1\\
	| \partial_y u | ~&\leq~ c_2 \ .
\end{align*}
The domain is divided equidistantly into $N_1 \times N_2$ pixels. Now we make a transition to the digital version $U \in \mathbb{R}^{N_1 \times N_2}$ of the image (in the sense of piecewise constant interpolation).
1. Derive bounds for the maximal difference of neighboring pixels in $U$.

Suppose $u_2$ is a pixel in the neighbourhood of $u_1$ and the distance between 2 pixels is $h$. By Taylor
$$
    u_2 = u_1 + \frac{\partial u_2}{\partial x}h+\frac{\partial u_2}{\partial y}h+O(h^2)
$$
Therefore
$$
    \|u_1-u_2\| = \| u_1 - u_1 + \frac{\partial u_2}{\partial x}h+\frac{\partial u_2}{\partial y}h+O(h^2) \| \\
    = \|\frac{\partial u_2}{\partial x}h+\frac{\partial u_2}{\partial y}h+O(h^2) \| \\
    \leq \|\frac{\partial u_2}{\partial x}h+\frac{\partial u_2}{\partial y}h\|+\|O(h^2) \| \\
    \leq (c_1+c_2)h + \|O(h^2)\|
$$

```python
import cv2
import numpy as np
alpha = 0.1

def lin_filter(image):
    kernel = np.array([
        [0, alpha, 0], 
        [alpha, 1-4*alpha, alpha], 
        [0, alpha, 0]]
    )
    return cv2.filter2D(image, -1, kernel)

sk.io.imshow(image)
sk.io.show()
sk.io.imshow(image_noise_poison)
sk.io.show()
sk.io.imshow(lin_filter(image_noise_poison))
sk.io.show()
```

## Deconvolution with Fourier transform

1. Specify a convolution kernel, apply a convolution to an image and add noise.
2. Using the direct (Fourier) deconvolution formula from the lecture deconvolve the image. What are the effects of the ill-posedness for increasing noise levels?
3. Compare your findings with the existing regularized deconvolution codes in scikit-image:

Richardson-Lucy deconvolution:
https://scikit-image.org/docs/dev/auto_examples/filters/plot_deconvolution.html?highlight=deconvolution

Wiener filter-based deconvolution:
https://scikit-image.org/docs/dev/auto_examples/filters/plot_restoration.html?highlight=deconvolution

```python
import scipy 

def convolution(image, kernel):
    return cv2.filter2D(image, -1, kernel)

def convolution2(image, kernel):
    return scipy.signal.convolve2d(image, kernel, mode="full", boundary="fill", fillvalue=0)

def noisify(image):
    return sk.util.random_noise(image, "gaussian")

def fourier_deconvolute(image, kernel):
    n = 2
    fourier_image = np.fft.fft2(image)
    fourier_kernel = np.fft.fft2(kernel, s=np.shape(image))    
    coefficents = np.divide(fourier_image, fourier_kernel)
    constant = 100*(2*np.pi)**(-n/2)
    return constant*np.abs(np.fft.ifft2(coefficents))
```

```python
kernel = np.array([
    [-3, -2, -1], 
    [-6, 1, 6], 
    [1, 2, 2]]
)
kernel_pascal = np.array([
    [0, 1, 2], 
    [1, 5, 0], 
    [4, 2, 1]]
)

kernel = kernel_pascal
image = sk.io.imread("Pascals_burger.jpeg")
gray_image = sk.color.rgb2gray(image)
noisy_image = noisify(gray_image)

print("Original")
sk.io.imshow(image)
sk.io.show()
print("Gray")
sk.io.imshow(gray_image, cmap="gray")
sk.io.show()
print("Noisy gray")
sk.io.imshow(noisy_image, cmap="gray")
sk.io.show()
```

## Convolution -> Deconvolution

```python
gray_image = sk.color.rgb2gray(image)
convulted_image = convolution2(gray_image, kernel)
deconvulted_convulted_image = fourier_deconvolute(convulted_image, kernel)

print("Gray")
sk.io.imshow(gray_image, cmap="gray")
sk.io.show()
print("Convoluted")
sk.io.imshow(convulted_image, cmap="gray")
sk.io.show()
print("Deconvoluted Convoluted")
sk.io.imshow(deconvulted_convulted_image, cmap="gray")
sk.io.show()
```

## Convolution -> Noisify -> Deconvolution

```python
convulted_image = convolution2(gray_image, kernel)
noisy_convulted_image = noisify(convulted_image)
deconvulted_noisy_convulted_image = fourier_deconvolute(noisy_convulted_image, kernel)

print("Convoluted")
sk.io.imshow(convulted_image, cmap="gray")
sk.io.show()
print("Noisy Convoluted")
sk.io.imshow(noisy_convulted_image, cmap="gray")
sk.io.show()
print("Deconvoluted Noisy Convoluted")
sk.io.imshow(deconvulted_noisy_convulted_image, cmap="gray")
sk.io.show()
```

## Convolution -> Noisify -> Richard

```python
convulted_image = convolution2(gray_image, kernel)
noisy_convulted_image = noisify(convulted_image)
richard_noisy_convulted_image = sk.restoration.richardson_lucy(noisy_convulted_image, kernel, iterations=30)

print("Convoluted")
sk.io.imshow(convulted_image, cmap="gray")
sk.io.show()
print("Noisy Convoluted")
sk.io.imshow(noisy_convulted_image, cmap="gray")
sk.io.show()
print("Richard deconvoluted")
sk.io.imshow(richard_noisy_convulted_image, cmap="gray")
sk.io.show()
```

## Convolution -> Noisify -> Wiener

```python
convulted_image = convolution2(gray_image, kernel)
noisy_convulted_image = noisify(convulted_image)
wiener_noisy_convulted_image = sk.restoration.unsupervised_wiener(noisy_convulted_image, kernel)

print("Convoluted")
sk.io.imshow(convulted_image, cmap="gray")
sk.io.show()
print("Noisy Convoluted")
sk.io.imshow(noisy_convulted_image, cmap="gray")
sk.io.show()
print("Richard deconvoluted")
sk.io.imshow(wiener_noisy_convulted_image, cmap="gray")
sk.io.show()
```
