## Convolution on a finite interval

We can similary define convolution with a Gaussian kernel on a finite interval $[0,\pi]$ through the initial boundary-value problem
$$v_t = v_{xx}, \quad v(t,0) = v(t,\pi) = 0,\quad v(0,x) = u(x)$$
with $f(x) = v(1,x)$.

The solution of the initial boundary-value problem is given by 
$$v(t,x) = \sum_{k=1}^{\infty} a_k\exp(- k^2 t)\sin(k x),$$
with $a_k$ are the Fourier sine coefficients of $u$:
$$a_k = \langle u, \sin(k\cdot) \rangle = \frac{2}{\pi}\int_0^{\pi} u(k) \sin (k x) \mathrm{d}x.$$

**1.** Define the forward operator $f = Ku$ in terms of the solution of the IBVP as $f(x) = v(1,x)$. Give the singular system of $K$, i.e., find $(\sigma_k, u_k, v_k)$ such that $Ku(x)$ can be expressed as 
$$Ku(x) = \sum_{k=0}^\infty \sigma_k \langle u, v_k \rangle u_k(x).$$


## Exercise 1
$$
f(x) = v(1,x) \\ 
    = \sum_{k=1}^{\infty} a_k exp(-k^2)sin(k x) \\
    = \sum_{k=1}^{\infty}exp(-k^2)\frac{2}{\pi}\int_0^1 u(x)\sin (k x) \mathrm{d}x\sin(k x) \\
    = \sum_{k=1}^{\infty}exp(-k^2)\langle u, \sin(k\cdot) \rangle\sin(k x) \\
$$
Suppose we define
$$
\sigma_k = \exp(-k^2) \\
v_k = \sin (k x)  \\
u_k = \sin(k x) \\
$$
Then
$$
Ku(x) = \sum_{k=0}^\infty \sigma_k \langle u, v_k \rangle u_k(x)
$$

**2.** The pseudo-inverse of $K$ is defined as
$$K^\dagger f(x) = \sum_{k=0}^n \sigma_k^{-1}\langle f, u_k \rangle v_k(x).$$

We can define a *regularized* pseudo-inverse through the variational problem
$$\min_{u} \|Ku - f\|^2 + \alpha R(u).$$
We investigate two types of regularization 
1. $$R(x) = \|u\|^2,$$
2. $$R(x) = \|u'\|^2.$$

Show that these lead to the following regularized pseudo-inverses
1. $$K_{\alpha}^\dagger f = \sum_{k=0}^\infty \frac{1}{\sigma_k + \alpha\sigma_k^{-1}}\langle f, u_k \rangle v_k(x).$$

2. $$K_{\alpha}^\dagger f = \sum_{k=0}^\infty \frac{1}{\sigma_k + \alpha k^2\sigma_k^{-1}}\langle f, u_k \rangle v_k(x)$$

**hint:** you can use the fact that the $v_k$ form an orthonormal basis for functions on $[0,1]$ and hence express the solution in terms of this basis.

## Exercise 2

### Exercise 1
$$
\min_{u} P := \min_{u} \|Ku - f\|^2 + \alpha \|u\|^2 = \min_{u} \langle Ku - f, Ku-f \rangle + \alpha \langle u,u\rangle
$$
The minimum is achieved when differention to u equals zero, thus
$$
0 = \frac{\partial P }{\partial u} \\
= \frac{\partial }{\partial u}\langle Ku - f, Ku-f \rangle + \alpha \langle u,u\rangle \\
\frac{\partial }{\partial u}\langle Ku, Ku \rangle - \langle f, Ku \rangle+ \langle f, f \rangle + \alpha \langle u,u\rangle \\
 = 2(K^TKu-f^TK + \alpha u)
$$
This means that
$$
u = K^\dagger f = (K^TK+\alpha)^{-1}K^Tf \\
 = (\Sigma^2_k+\alpha I)^{-1}\sum_{k=0}^\infty\sigma_k\langle f, u_k\rangle v_k \\
 = \sum_{k=0}^\infty\frac{\sigma_k}{\sigma^2_k+\alpha}\langle f, u_k\rangle v_k \\
$$
where we used that $v_k$ are an orthogonal basis and that 
$$
 K^Tf = \sum_{k=0}^\infty \sigma_k \langle f, u_k \rangle v_k(x)
$$

### Exercise 2
$$
\min_{u} P := \min_{u} \|Ku - f\|^2 + \alpha \|u\|^2 = \min_{u} \langle Ku - f, Ku-f \rangle + \alpha \langle u',u'\rangle
$$
The minimum is achieved when differention to u equals zero, thus
$$
0 = \frac{\partial P }{\partial u} \\
= \frac{\partial }{\partial u}\langle Ku - f, Ku-f \rangle + \alpha \langle u',u'\rangle \\
\frac{\partial }{\partial u}\langle Ku, Ku \rangle - \langle f, Ku \rangle+ \langle f, f \rangle + \alpha \langle u',u' \rangle \\
 = 2(K^TKu-f^TK + \alpha \frac{\partial \langle u',u' \rangle}{\partial u})
$$
From the definition of $u$ we know that 
$$
u'= \frac{\partial u }{\partial x} \\
= \frac{\partial \sum_{k=1}^{\infty} a_k\sin(k x)}{\partial x} \\
= \sum_{k=1}^{\infty} a_kk\cos(k x)
$$
Therefore
$$
\langle u',u' \rangle 
= \langle \sum_{k=1}^{\infty} a_kk\cos(k x),\sum_{k=1}^{\infty} a_kk\cos(k x) \rangle \\
= \sum_{k=1}^{\infty}\sum_{n=1}^{\infty} \langle a_kk\cos(k x), a_nn\cos(n x) \rangle \\
=  \sum_{k=1}^{\infty}a^2_kk^2\langle \cos(k x),\cos(k x) \rangle \\
=  \sum_{k=1}^{\infty}a^2_kk^2\int_0^\pi \cos(k x)\cos(k x)dx \\
=  \sum_{k=1}^{\infty}a^2_kk^2\int_0^\pi \sin(k x)\sin(k x)dx \quad \text{integration by parts} \\
=  \int_0^\pi \sum_{k=1}^{\infty}a^2_kk^2\sin(k x)\sin(k x)dx \\
= \langle \sum_{k=1}^{\infty} a_kk\sin(k x),\sum_{k=1}^{\infty} a_kk\sin(k x) \rangle
$$
This implies
$$
\frac{\partial \langle u',u' \rangle}{\partial u}=2\sum_{k=1}^{\infty} a_kk^2\sin(k x) 
$$
Substituting this and the SVD of K gives
$$
0 = K^TKu-f^TK + \alpha \frac{\partial\langle u',u' \rangle}{\partial u}\\
= \sum_{k=1}^{\infty}\sigma^2_ka_k\sin(k x)-\sum_{k=0}^\infty\sigma_k\langle f, u_k\rangle v_k+\alpha\sum_{k=1}^{\infty} a_kk^2\sin(k x) \\
= \sum_{k=1}^{\infty}(\sigma^2_k+\alpha k^2)a_k\sin(k x)-\sum_{k=0}^\infty\sigma_k\langle f, u_k\rangle v_k \\
= \sum_{k=1}^{\infty}(\sigma^2_k+\alpha k^2)\langle u, u_k\rangle v_k-\sum_{k=0}^\infty\sigma_k\langle f, u_k\rangle v_k \\
$$
$v_k$ are orthonormal and $(\sigma^2_k-\alpha k^2)\neq0$, hence
$$
\sum_{k=1}^{\infty}\langle u, u_k\rangle v_k=\sum_{k=0}^\infty\frac{\sigma_k}{\sigma^2_k+\alpha k^2}\langle f, u_k\rangle v_k =\sum_{k=0}^\infty\frac{1}{\sigma_k+\alpha k^2\sigma^{-1}_k}\langle f, u_k\rangle v_k \\
$$
Therefore
$$
K^\dagger f = u = \sum_{k=1}^{\infty}\langle u, u_k\rangle v_k=\sum_{k=0}^\infty\frac{1}{\sigma_k+\alpha k^2\sigma^{-1}_k}\langle f, u_k\rangle v_k \\
$$



**3.** We can now study the need for regularization, assuming that the Fourier coefficients $f_k = \langle f, u_k \rangle$ of $f$ are given. Determine which type of regularization (if any) is needed to satisty the Picard condition in the following cases (you can set $\alpha = 1$ for this analysis)

1. $f_k = \exp(-2 k^2)$
2. $f_k = k^{-1}$

## Exercise 3

The picard condition holds iff
$$
\sum_{k=0}^\infty \frac{|\langle f, u_k \rangle|^2}{\sigma^2_k} < \infty
$$
From exercise 1 we know that 
$$
\sigma_k = exp(-k^2)
$$

### $f_k = \exp(-2 k^2)$

$$
\sum_{k=0}^\infty \frac{|\langle f, u_k \rangle|^2}{\sigma^2_k} = \frac{\pi^2}{4}\sum_{k=0}^\infty \frac{(e^{-2k^2})^2}{e^{-2k^2}} \\
= \sum_{k=0}^\infty \frac{e^{-4k^2}}{e^{-2k^2}} \\
= \sum_{k=0}^\infty e^{-2k^2} \\
$$
Using the ratio test
$$
\lim_{k\to\infty}|\frac{e^{-2(k+1)^2}}{e^{-2k^2}}| = \lim_{k\to\infty}|\frac{e^{-2(k^2+2k+1)}}{e^{-2k^2}}| \\
= \lim_{k\to\infty}|\frac{e^{-2k^2-4k-2)}}{e^{-2k^2}}| \\
= \lim_{k\to\infty}|e^{-4k-2)}|<1 \\
$$
we have that the Picard condition holds.

### $f_k = k^{-1}$

#### without regularization
$$
\sum_{k=0}^\infty \frac{|\langle f, u_k \rangle|^2}{\sigma^2_k} = \sum_{k=0}^\infty \frac{e^{4k^2}}{k^2} =\infty
$$
#### method 1
Let us try regularization method 1.
$$
\sum_{k=0}^\infty \frac{|\langle f, u_k \rangle|^2}{(\sigma_k+\sigma^{-1}_k)^2}
=\sum_{k=0}^\infty \frac{k^{-2}}{(e^{-k^2}+e^{k^2})^2} \\
=\sum_{k=0}^\infty \frac{k^{-2}}{e^{-2k^2}+2+e^{2k^2}} \\
=\sum_{k=0}^\infty \frac{1}{k^2(e^{-2k^2}+2+e^{2k^2})} \\
$$
By the ratio test this converges.
#### method 2
Let us try regularization method 2.
$$
\sum_{k=0}^\infty \frac{|\langle f, u_k \rangle|^2}{(\sigma_k+k^2\sigma^{-1}_k)^2}=\sum_{k=0}^\infty \frac{k^{-2}}{(e^{-k^2}+k^2e^{k^2})^2} \\
=\sum_{k=0}^\infty \frac{k^{-2}}{e^{-2k^2}+2k^2+k^4e^{2k^2}} \\
=\sum_{k=0}^\infty \frac{1}{k^2(e^{-2k^2}+2k^2+k^4e^{2k^2})} \\
=|\sum_{k=0}^\infty \frac{1}{k^2(e^{-2k^2}+2k^2+k^4e^{2k^2})}| \\
\leq\sqrt{\sum_{k=0}^\infty \frac{1}{k^4}}+\sqrt{\sum_{k=0}^\infty \frac{1}{(e^{-2k^2}+2k^2+k^4e^{2k^2})^2}}<\infty \\
$$

**4.** Compute the bias and variance for $u(x) = \sin(k x)$ and  measurements $f^{\delta}(x) = Ku(x) + \delta \sin(\ell x)$ for fixed $k < \ell$ and $\delta$. Plot the bias and variance for well-chosen $k,\ell$ and $\delta$ and discuss the difference between the two types of regularization.

### Exercise 4

The bias and variance are defined to be
$$
bias := \|(K^\dagger - K_\alpha^\dagger)f)\| \\
variance := \|K_\alpha^\dagger(f - f^\delta)\|
$$

We know 
$$
Ku(x) = \sum_{k=0}^\infty \sigma_k \langle u, v_k \rangle u_k(x)
$$
with
$$
\sigma_k = \exp(-k^2) \\
v_k = \sin (k x)  \\
u_k = \sin(k x) \\
$$
This means that if $u(x) = \sin(k_0x)$, then
$$
f = Ku(x) \\
= \sum_{k=0}^\infty \sigma_k \langle \sin(k_0x), \sin(kx) \rangle \sin(kx) \\
= \sigma_{k_0}\sin(k_0x) \\
$$
and 
$$
f^\delta = Ku(x)+\delta\sin(\ell x)\\
= \sigma_{k_0}\sin(k_0x)+\delta\sin(\ell x)
$$
Therefore
$$
K^\dagger f = u \\
K_{1,a}^\dagger f = \frac{\sigma^2_{k_0}}{\sigma^2_{k_0}+\alpha}u \\
K_{2,a}^\dagger f = \frac{\sigma^2_{k_0}}{\sigma^2_{k_0}+\alpha k^2_0}u \\
K_{1,a}^\dagger f^\delta = \frac{\sigma_{k_0}}{\sigma^2_{k_0}+\alpha}\sigma_{k_0}u+\frac{\sigma_{l}}{\sigma^2_{l}+\alpha}\delta \sin(\ell x) \\
K_{2,a}^\dagger f^\delta = \frac{\sigma_{k_0}}{\sigma^2_{k_0}+\alpha k^2_0}\sigma_{k_0}u+\frac{\sigma_{l}}{\sigma^2_{l}+\ell^2\alpha}\delta \sin(\ell x) \\
$$
Thus the bias and variance are for regularization method 1
$$
bias_1 := \|(K^\dagger - K_{1,\alpha}^\dagger)f)\| = \bigg\|\bigg(1 - \frac{\sigma^2_{k_0}}{\sigma^2_{k_0}+\alpha}\bigg)u)\bigg\| = \bigg\|\frac{\alpha}{\sigma^2_{k_0}+\alpha}u\bigg\| = \bigg|{\frac{\alpha}{\sigma^2_{k_0}+\alpha}}\bigg|\\
variance_1 := \|K_{1,\alpha}^\dagger(f - f^\delta)\|= \bigg\|\frac{\sigma_{l}}{\sigma^2_{l}+\alpha}\delta \sin(\ell x)\bigg\|=\bigg|\frac{\sigma_{l}}{\sigma^2_{l}+\alpha}\delta\bigg|
$$
and the bias and variance are for regularization method 2
$$
bias_2 := \|(K^\dagger - K_{2,\alpha}^\dagger)f)\| = \bigg\|\bigg(1 - \frac{\sigma^2_{k_0}}{\sigma^2_{k_0}+k_0^2\alpha}\bigg)u)\bigg\| = \bigg\|\frac{\alpha k^2_0}{\sigma^2_{k_0}+k_0^2\alpha}u\bigg\|= \bigg|\frac{\alpha k^2_0}{\sigma^2_{k_0}+k_0^2\alpha}\bigg| \\
variance_2 := \|K_{2,\alpha}^\dagger(f - f^\delta)\|= \bigg\|\frac{\sigma_{l}}{\sigma^2_{l}+\ell^2\alpha}\delta \sin(\ell x)\bigg\|= \bigg|\frac{\sigma_{l}}{\sigma^2_{l}+\ell^2\alpha}\delta\bigg|
$$

```python
k = 2
ell = 3
alpha = 1
delta = 0.1
n = 100
x = np.linspace(0,1,n)
kaas = np.logspace(0,20,base=2)
kaas2 = np.linspace(0,5)
kaas3 = np.linspace(0,20)
ells = np.linspace(0,25)
deltas = np.logspace(0,-18,36)

def sigma(k):
    return np.exp(-k**2)

def getU(k,c=1):
    u = c*np.sin(k*x)
    return u

def bias(c, k, alpha=0, type=None):
    bias = float()
    if type is None:
        bias = 0
    elif type==1:
        bias = np.abs(alpha/(sigma(k)**2+alpha)*c)
    elif type==2:
        bias = np.abs(k**2*alpha/(sigma(k)**2+k**2*alpha)*c)
    else:
        print("Invalid type")
        bias = None
    return bias

def variance(noise, ell, alpha=0, type=None):
    variance = float()
    if type is None:
        variance = 0
    elif type==1:
        variance = np.linalg.norm(sigma(ell)/(sigma(ell)**2+alpha)*noise)
    elif type==2:
        variance = np.linalg.norm(sigma(ell)/(sigma(ell)**2+ell**2*alpha)*noise)
    else:
        print("Invalid type")
        variance = None
    return variance
```

```python
variances_1 = [variance(delta,ell,1,type=1) for delta in deltas]
variances_2 = [variance(delta,ell,1,type=2) for delta in deltas]

plt.figure()
plt.title("Linearity of variance to noise amplitude -- logarithmic")
plt.semilogy(variances_1, label="regu 1")
plt.semilogy(variances_2, label="regu 2")
plt.ylabel("Variance")
plt.xlabel("Twice the log of noise amplitude")
plt.legend()

plt.figure()
plt.title("Linearity of variance to noise amplitude")
plt.plot(deltas,variances_1, label="regu 1")
plt.plot(deltas,variances_2, label="regu 2")
plt.ylabel("Variance")
plt.xlabel("Noise amplitude")
plt.legend()

bias_1 = [bias(1,k,1,type=1) for k in kaas]
bias_2 = [bias(1,k,1,type=2) for k in kaas]

plt.figure()
plt.title("Bias for varying k")
plt.plot(kaas, bias_1, label="regu 1")
plt.plot(kaas, bias_2, label="regu 2")
plt.ylabel("bias")
plt.xlabel("k")
plt.legend()

bias_1 = [bias(1,k,1,type=1) for k in kaas2]
bias_2 = [bias(1,k,1,type=2) for k in kaas2]

plt.figure()
plt.title("Bias for varying k")
plt.plot(kaas2, bias_1, label="regu 1")
plt.plot(kaas2, bias_2, label="regu 2")
plt.ylabel("bias")
plt.xlabel("k")
plt.legend()

variances_1L = [variance(delta,ell,1,type=1) for ell in ells]
variances_2L = [variance(delta,ell,1,type=2) for ell in ells]

plt.figure()
plt.title("Variance for varying l")
plt.semilogy(ells, variances_1L, label="regu 1")
plt.semilogy(ells, variances_2L, label="regu 2")
plt.ylabel("variance")
plt.xlabel("ell")
plt.legend()

plt.show()
```

```python
kaas3 = np.linspace(0,20)
alphas = np.logspace(0,-5,5)
plt.figure()
for alpha in alphas:
    variances_1 = [variance(delta,ell,alpha=alpha,type=1) for delta in deltas]
    plt.title("Variance for various alpha")
    plt.loglog(deltas,variances_1, label="alpha:{:3f}".format(alpha))
    plt.legend()
plt.ylabel("variance")
plt.figure()
for alpha in alphas:
    variances_2 = [variance(delta,ell,alpha=alpha,type=2) for delta in deltas]
    plt.title("Variance for various alpha")
    plt.loglog(deltas,variances_2, label="alpha:{:3f}".format(alpha))
plt.ylabel("variance")
plt.legend()

plt.figure()
plt.title("Bias for varying alpha: regu 1")
for alpha in alphas:
    biases_1 = [bias(1,k,alpha=alpha,type=1) for k in kaas3]
    plt.plot(kaas3, biases_1, label="alpha:{:3f}".format(alpha))
plt.ylabel("bias")
plt.xlim(-.1,10)
plt.legend()

plt.figure()
plt.title("Bias for varying alpha: regu 2")
for alpha in alphas:
    biases_2 = [bias(1,k,alpha=alpha,type=2) for k in kaas3]
    plt.plot(kaas3,biases_2, label="alpha:{:3f}".format(alpha))
plt.xlim(-.1,10)
plt.ylabel("bias")
plt.legend()

plt.show()
```

```python
ell = 10
delta = 0.1
bias_1_L = [bias(1,k,1,type=1) for k in range(ell)]
bias_2_L = [bias(1,k,1,type=2) for k in range(ell)]
plt.figure()
plt.title("")
plt.ylabel("bias")
plt.xlabel("k")
plt.plot(bias_1_L)
plt.plot(bias_2_L)    
```

### Explanation


We have that the sum of the bias and variance is a measure for how good the regularized method on noisy signal approaches the noiseless situation.
$$
\|K^\dagger f-K^\dagger_\alpha f^\delta\|\leq \|(K^\dagger-K^\dagger_\alpha)f \|+\|K_\alpha^\dagger(f-f^\delta)\| = bias +variance
$$

We see that regardless of the regularization method, that the variance decreases exponentially with $\ell$. We also see that the bias goes to one slower for regularization method 2 than regularization method 1. For suffiently low $k$ both have a low bias, but the bias of regularization method 2 is lower. Therefore for small $k$, positive $\alpha$, a sufficiently large $\ell$ both methods have low bias and variance, but regularization method 2 has lower bias and variance.
