## Exercise 1

Let $\Omega \subset \mathbb{R}^2$ and $\Sigma \subset \mathbb{R}^2$. Compute the Fréchet derivatives of the following functionals:

* $J(u) = \frac{1}{2} \left\| \nabla u \right\|_{L^2(\Omega)}^2$ where $u \in W^{1,2}(\Omega)$.
* $J(u) = \frac{1}{2} \left\| Ku-f \right\|_{L^2(\Sigma)}^2$ where $K: L^2(\Omega) \rightarrow L^2(\Sigma)$ is a compact linear operator, $u : \Omega \rightarrow \mathbb{R}$ and $f : \Sigma \rightarrow \mathbb{R}$.
* $J(\mathbf{v}) = \frac{1}{2} \left\| \partial_t f + \nabla\cdot(f \mathbf{v}) \right\|_{L^2(\Omega \times [0,T])}^2$ 
	where $f$ here represents an image sequence, i.e. $f: \Omega \times [0,T] \rightarrow \mathbb{R}$, and $\mathbf{v}$ denotes a desired vector field, i.e. $\mathbf{v}: \Omega \times [0,T] \rightarrow \mathbb{R}^2$.


Let $V,W$ be normed vector spaces and $\Omega\subset V$. A function $f:\Omega\to W$ is called Frèchlet differentiable at $x\in\Omega$ if there exists a bounded, linear operator $A:V\to W$ such that
$$
\lim_{h\to0}\frac{\|f(x+h)-f(x)-Ah\|_W}{\|h\|_V} = 0
$$

### Solution 1

Let
$$
J(u) = \frac{1}{2} \left\| \nabla u \right\|_{L^2(\Omega)}^2
$$
then
$$
\begin{align*}
    W &= L^2(\mathbb{R}) \\
    V &= W^{1,2}(\mathbb{R})
\end{align*}
$$
Starting 
$$
\begin{align*}
    J(u+v) &= \frac{1}{2} \left\| \nabla (u+v) \right\|_{L^2(\Omega)}^2 \\
           &= \frac{1}{2} \left\| \nabla u +\nabla v \right\|_{L^2(\Omega)}^2 \\
           &= \frac{1}{2} \left\| \nabla u \right\|_{L^2(\Omega)}^2+\frac{1}{2} \left\| \nabla u \right\|_{L^2(\Omega)}^2+\langle \nabla u | \nabla v \rangle_{L^2(\Omega)} \\
           &= J(u)+J(v)+\langle \nabla u | \nabla v \rangle_{L^2(\Omega)} \\
\end{align*}
$$
Hence
$$
\begin{align*}
\lim_{\|v\|\to0}\frac{\|J(u+v)-J(u)-Av\|_W}{\|v\|_V} &= \lim_{\|v\|\to0}\frac{\|J(u)+J(v)+\langle \nabla u | \nabla v \rangle_{L^2(\Omega)}-J(u)-Av\|_W}{\|v\|_V} \\
&= \lim_{\|v\|\to0}\frac{\|J(v)+\langle \nabla u | \nabla v \rangle_{L^2(\Omega)}-Av\|_W}{\|v\|_V} \\
&\leq \lim_{\|v\|\to0}\frac{\|J(v)\|_W}{\|v\|_V}+\frac{\|\langle \nabla u | \nabla v \rangle_{L^2(\Omega)}-Av\|_W}{\|v\|_V} \\
\end{align*}
$$
By Poincaré
$$
\begin{align*}
    \lim_{\|v\|\to 0}\frac{\|J(v\|_W}{\|v\|_V} 
        &= \lim_{\|v\|\to 0}\frac{\|\frac{1}{2} \left\| \nabla v \right\|_{L^2(\Omega)}^2\|_W}{\|v\|_V} \\
        &= \lim_{\|v\|\to 0}\frac{\|\frac{1}{2} \left\| \nabla v \right\|_{L^2(\Omega)}^2\|_{L^2(\mathbb{R})}}{\|v\|_{W^{1,2}(\mathbb{R})}} \\
        &\leq C\lim_{\|v\|\to 0}\frac{\|v\|^2}{\|v\|}_{W^{1,2}(\Omega)} =0 
\end{align*}
$$
Let 
$$
Av = \langle \nabla u | \nabla v \rangle_{L^2(\Omega)}
$$
then this $A$ is linear bounded and solves the limit, thus this $A$ is the Frechlét derivative.

### Solution 2

Let
$$
J(u) = \frac{1}{2} \left\| Ku-f \right\|_{L^2(\Sigma)}^2
$$
then
$$
\begin{align*}
    W &= \mathbb{R} \\
    V &= L^2(\mathbb{R})
\end{align*}
$$
Starting 
$$
\begin{align*}
    J(u+v) &= \frac{1}{2} \left\| K(u+v)-f \right\|_{L^2(\Sigma)}^2 \\
           &= \frac{1}{2} \left\| Ku+Kv-f \right\|_{L^2(\Sigma)}^2 \\
           &= \frac{1}{2} \left\| Ku-f \right\|_{L^2(\Sigma)}^2+\frac{1}{2} \left\| Kv \right\|_{L^2(\Sigma)}^2+\langle Ku-f|Kv \rangle_{L^2(\Sigma)} \\
\end{align*}
$$
Hence
$$
\begin{align*}
\lim_{\|v\|\to0}\frac{\|J(u+v)-J(u)-Av\|_W}{\|v\|_V} 
&= \lim_{\|v\|\to0}\frac{\|\frac{1}{2} \left\| Ku-f \right\|_{L^2(\Sigma)}^2+\frac{1}{2} \left\| Kv \right\|_{L^2(\Sigma)}+\langle Ku-f|Kv \rangle_{L^2(\Sigma)}^2-J(u)-Av\|_W}{\|v\|_V} \\
&= \lim_{\|v\|\to0}\frac{\|\frac{1}{2} \left\| Kv \right\|_{L^2(\Sigma)}^2+\langle Ku-f|Kv \rangle_{L^2(\Sigma)}-Av\|_W}{\|v\|_V} \\
&\leq \lim_{\|v\|\to0}\frac{\frac{1}{2} \left\| Kv \right\|_{L^2(\Sigma)}^2}{\|v\|_V}+\frac{\|\langle Ku-f|Kv \rangle_{L^2(\Sigma)}-Av\|_W}{\|v\|_V} \\
\end{align*}
$$
$K$ is compact, thus
$$
\lim_{\|v\|\to 0}\frac{\|Kv\|^2}{\|v\|}_{L^2(\Omega)} \leq \lim_{\|v\|\to 0}\|K\|^2\|v\| =0
$$
Let 
$$
Av =\langle Ku-f|Kv \rangle_{L^2(\Sigma)}
$$
then this $A$ is linear bounded and solves the limit, thus this $A$ is the Frechlét derivative.

### Solution 3

Let
$$
J(u) =  \frac{1}{2} \left\| \partial_t f + \nabla\cdot(f u) \right\|_{L^2(\Omega \times [0,T])}^2
$$
then
$$
\begin{align*}
    W &= L^2(\mathbb{R}) \\
    V &= L^{2}(\mathbb{R} \times [0,T])
\end{align*}
$$
Starting 
$$
\begin{align*}
    J(u+v) &= \frac{1}{2} \left\| \partial_t f + \nabla\cdot(f (u+v) \right\|_{L^2(\Omega \times [0,T])}^2 \\
    &= \frac{1}{2} \left\| \partial_t f + \nabla\cdot fu+\nabla\cdot fv \right\|_{L^2(\Omega \times [0,T])}^2 \\
\end{align*}
$$
Hence
$$
\begin{align*}
\lim_{\|v\|\to0}\frac{\|J(u+v)-J(u)-Av\|_W}{\|v\|_V} 
    &= \lim_{\|v\|\to0}\frac{\|\frac{1}{2} \left\| \partial_t f + \nabla\cdot fu+\nabla\cdot fv \right\|_{L^2(\Omega \times [0,T])}^2-J(u)-Av\|_W}{\|v\|_V} \\
    &= \lim_{\|v\|\to0}\frac{\|\frac{1}{2} \left\| \partial_t f + \nabla\cdot fu+\nabla\cdot fv \right\|_{L^2(\Omega \times [0,T])}^2-\|\frac{1}{2} \left\| \partial_t f + \nabla\cdot fu\right\|_{L^2(\Omega \times [0,T])}^2-Av\|_W}{\|v\|_V} \\
    &= \lim_{\|v\|\to0}\frac{\|\frac{1}{2} \left\| \nabla\cdot fv \right\|_{L^2(\Omega \times [0,T])}^2+\langle \partial_t f+\nabla\cdot fu|\nabla\cdot fv \rangle_{L^2(\Omega\times[0,T])}-Av\|_W}{\|v\|_V} \\
    &\leq \lim_{\|v\|\to0}\frac{\|\frac{1}{2} \left\| \nabla\cdot fv \right\|_{L^2(\Omega \times [0,T])}^2}{\|v\|_V}+\frac{\|\langle \partial_t f+\nabla\cdot fu|\nabla\cdot fv \rangle_{L^2(\Omega\times[0,T])}-Av\|_W}{\|v\|_V} \\
\end{align*}
$$
Let 
$$
Av =\langle \partial_t f+\nabla\cdot fu|\nabla\cdot fv \rangle_{L^2(\Omega\times[0,T])}
$$
then this $A$ is linear bounded and solves the limit, thus this $A$ is the Frechlét derivative.

## Exercise 2

Consider the Tikhonov functional for denoising (in $\mathbb{R}^n$):
$$\min\limits_x\frac{1}{2}\left\|x-y\right\|_2^2 + \frac{\alpha}{2}\left\|x\right\|_2^2.$$

* Give the solution of this variational problem explicitly.
* Generate in Python a random 1x128 vector with 5 non-zero coefficients (entries) and add aussian noise with standard deviation $\sigma = 0.05$ (see example below)
* Denoise the vector by solving the variational problem. What happens for different regularisation parameters $\alpha = \left\{0.01, 0.05, 0.1, 0.2\right\}$?. Consider in particular $\alpha=0.1$. *Is the solution sparse?*

$$
\min\limits_u\frac{1}{2}\left\|u-v\right\|_2^2 + \frac{\alpha}{2}\left\|u\right\|_2^2  \\
\iff
\frac{\partial}{\partial u}\bigg(\frac{1}{2}\left\|u-v\right\|_2^2 + \frac{\alpha}{2}\left\|u\right\|_2^2\bigg) =0 \\
\iff
\frac{\partial}{\partial u}\bigg(\frac{1}{2}\left\|u\right\|_2^2-\langle u|v\rangle_2 + \frac{\alpha}{2}\left\|u\right\|_2^2\bigg) =0 \\
\\
\iff
\bigg(\langle u|\phi\rangle_2-\langle v|\phi\rangle_2 + \alpha\langle u|\phi \rangle_2\bigg) =0 \\
\\
\iff
\langle (\alpha+1)u-v|\phi\rangle_2 =0 \\
\\
\iff
(\alpha+1)u-v =0\\
\iff
u=\frac{v}{\alpha+1}
$$

```python
n = 128
k = 5
sigma = 0.05

x = np.zeros(n)
x[np.random.randint(128,size=k)]=np.random.randn(k)
y = x + sigma*np.random.randn(n)
plt.plot(x)
plt.plot(y)
```

```python
alphas = [0.01, 0.05, 0.1, 0.2]
for alpha in alphas:
    x_re = y/(alpha+1)
    plt.figure()
    plt.plot(x_re, label=r"\alpha={}".format(alpha))
    plt.legend()
```

$$
\begin{align*}
\min\limits_u\frac{1}{2}\left\|u-v\right\|_2^2 + \alpha\left\|u\right\|_1 
&\iff
\frac{\partial}{\partial u}\bigg(\frac{1}{2}\left\|u-v\right\|_2^2 + \alpha\left\|u\right\|_1\bigg) =0 \\
&\iff
\frac{\partial}{\partial u}\bigg(\frac{1}{2}\left\|u\right\|_2^2-\langle u|v\rangle_2+\alpha\left\|u\right\|_1\bigg) =0 \\
&\iff \forall k\in\mathbb{Z}^+:\quad\frac{\partial}{\partial u_k}\int_\Omega \frac{1}{2}\left(\sum_i u_i\phi_i\right)^2-\left(\sum_i u_i\phi_i\sum_i v_i\phi_i\right)+\alpha\left|\sum_i u_i\phi_i\right|d\Omega = 0
\\
&\iff \forall k\in\mathbb{Z}^+:\quad\int_\Omega \left(\sum_i u_i\phi_i\right)\phi_k-\left(\sum_i v_i\phi_i\right)\phi_k+\alpha sign\left\{\sum_i u_i\phi_i\right\}\phi_kd\Omega = 0
\\
&\iff
\bigg(\langle u|\phi\rangle_2-\langle v|\phi\rangle_2 + \alpha\langle sign\{u\}|\phi \rangle_2\bigg) =0 \\
\\
&\iff
\langle u+\alpha sign\{u\}-v|\phi\rangle_2 =0 \\
\\
&\iff
u+\alpha sign\{u\}-v =0\\
&\iff
u+\alpha sign\{u\}=v\\
\end{align*}
$$
